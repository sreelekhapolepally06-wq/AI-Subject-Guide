```python
from google import genai
from google.colab import userdata


api_key = userdata.get('GEMINI_API_KEY')

if not api_key:
    raise ValueError("API key not found in Secrets.")


client = genai.Client(api_key=api_key)

print("API Loaded Successfully!")
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Hello"
)

print(response.text)
```

```python
import PyPDF2

file_name = list(uploaded.keys())[0]
document_text = ""

with open(file_name, "rb") as f:
    reader = PyPDF2.PdfReader(f)
    for i, page in enumerate(reader.pages):
        text = page.extract_text()
        if text:
            document_text += text + "\n"
        else:
            print(f"Warning: Page {i+1} has no text.")

print("PDF Loaded Successfully!")
print(f"Total characters extracted: {len(document_text)}")
```

```python
print(document_text[:500])
```

```python
print("Ask questions from your PDF. Type 'exit' to stop.\n")

while True:
    question = input("You: ")

    if question.lower() == "exit":
        break

    prompt = f"""
    Answer the question ONLY using the document below.
    If the answer is not in the document, say 'Answer not found in document.'

    Document:
    {document_text}

    Question:
    {question}
    """

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    print("Bot:", response.text)
```
